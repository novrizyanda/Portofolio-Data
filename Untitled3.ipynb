```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta

import matplotlib.pyplot as plt
import matplotlib
import seaborn as sns
```

```python
data = pd.read_csv('ecommerce.csv')
data
```

```python
df_data = data.copy()
df_data
```

```python
def data_information(df_data):
    print(df_data.info())
    print(df_data.isnull().sum())
    print("Order ID Has ", df_data[df_data[['Order ID']].duplicated()].shape[0], " duplicated rows.")
    print("Customer ID Has ", df_data[df_data[['Customer ID']].duplicated()].shape[0], " duplicated rows.")
#     print(df_data.groupby(df_data[['Ship Mode']].columns.tolist(),as_index=False).size())

data_information(df_data)
```

Data Cleaning: Handling Null Data, Converting Data Types, Adding New Column

Handling Null Data

find row with null data

```python
data_null = df_data[df_data.isnull().any(axis=1)]
data_null
```

Fill Null data

```python
df_data = df_data.fillna({
    'Order Date' : '2015-04-17',
#     calculate aging vs ship date
    'Aging' : 6,
#     calculate order date vs ship date
#     sales, quantity, discount, shipping cost. we don't have information related those metrics. the data will be null
    'Order Priority' : 'NaN',
    'Customer ID' : 'NaN',
    'Segment' : 'NaN',
    'Region' : 'South'
    
})
```

Handing Date Data

```python
df_data[['Order Date','Ship Date']].head(10)
```

Convert String to Date

```python
df_data['Order Date'] = pd.to_datetime(df_data['Order Date'])
df_data['Ship Date'] = pd.to_datetime(df_data['Ship Date'])
df_data[['Order Date','Ship Date']].head(10)
```

Handing Float Data

```python
df_data[['Sales','Quantity','Discount','Profit','Shipping Cost']].head(10)
```

Remove $ icon and convert to float type

```python
df_data['Sales'] = df_data['Sales'].str.strip('$')
df_data['Sales'] = pd.to_numeric(df_data['Sales'], errors='coerce')
df_data['Quantity'] = pd.to_numeric(df_data['Quantity'], errors='coerce')
df_data['Discount'] = pd.to_numeric(df_data['Discount'], errors='coerce')
df_data['Profit'] = df_data['Profit'].str.strip('$')
df_data['Profit'] = pd.to_numeric(df_data['Profit'], errors='coerce')
df_data['Shipping Cost'] = df_data['Shipping Cost'].str.strip('$')
df_data['Shipping Cost'] = pd.to_numeric(df_data['Shipping Cost'], errors='coerce')
df_data[['Sales','Quantity','Discount','Profit','Shipping Cost']].head(10)
```

Adding New Column || Adding Cost/Quantity & Year-Month

```python
df_data['sales_per_qty'] = df_data['Sales']/df_data['Quantity']
df_data['sales_per_qty'] = pd.to_numeric(df_data['sales_per_qty'], errors='coerce')
df_data['discount_per_qty'] = df_data['Discount']/df_data['Quantity']
df_data['discount_per_qty'] = pd.to_numeric(df_data['discount_per_qty'], errors='coerce')
df_data['profit_per_qty'] = df_data['Profit']/df_data['Quantity']
df_data['profit_per_qty'] = pd.to_numeric(df_data['profit_per_qty'], errors='coerce')
df_data['shipping_cost_per_qty'] = df_data['Shipping Cost']/df_data['Quantity']
df_data['shipping_cost_per_qty'] = pd.to_numeric(df_data['shipping_cost_per_qty'], errors='coerce')
df_data['year_month'] = df_data['Order Date'].dt.to_period('M')
df_data
```

```python
shipmode_count_order = pd.pivot_table(df_data, values='Order ID', index=['Ship Mode'], aggfunc=np.count_nonzero).reset_index()
shipmode_qty = pd.pivot_table(df_data, values='Quantity', index=['Ship Mode'], aggfunc=np.sum).reset_index()
# 
shipmode_sales = pd.pivot_table(df_data, values='Sales', index=['Ship Mode'], aggfunc=np.sum).reset_index()
shipmode_avg_sales = pd.pivot_table(df_data, values='Sales', index=['Ship Mode'], aggfunc=np.mean).reset_index()
# 
shipmode_profit = pd.pivot_table(df_data, values='Profit', index=['Ship Mode'], aggfunc=np.sum).reset_index()
shipmode_avg_profit = pd.pivot_table(df_data, values='Profit', index=['Ship Mode'], aggfunc=np.mean).reset_index()
# 
shipmode_ship_cost = pd.pivot_table(df_data, values='Shipping Cost', index=['Ship Mode'], aggfunc=np.sum).reset_index()
shipmode_avg_ship_cost = pd.pivot_table(df_data, values='Shipping Cost', index=['Ship Mode'], aggfunc=np.mean).reset_index()
# 
shipmode_avg_aging = pd.pivot_table(df_data, values='Aging', index=['Ship Mode'], aggfunc=np.mean).reset_index()
# 
shipmode_avg_sales_per_qty = pd.pivot_table(df_data, values='sales_per_qty', index=['Ship Mode'], aggfunc=np.mean).reset_index()
shipmode_avg_profit_per_qty = pd.pivot_table(df_data, values='profit_per_qty', index=['Ship Mode'], aggfunc=np.mean).reset_index()
shipmode_avg_shipping_cost_per_qty = pd.pivot_table(df_data, values='shipping_cost_per_qty', index=['Ship Mode'], aggfunc=np.mean).reset_index()

# Rename Columns
shipmode_count_order.columns = ['Ship Mode','Count Order']
shipmode_sales.columns = ['Ship Mode','sum_of_Sales']
shipmode_avg_sales.columns = ['Ship Mode','avg_of_Sales']
shipmode_profit.columns = ['Ship Mode','sum_of_Profit']
shipmode_avg_profit.columns = ['Ship Mode','avg_of_Profit']
shipmode_ship_cost.columns = ['Ship Mode','sum_of_Shipping Cost']
shipmode_avg_ship_cost.columns = ['Ship Mode','avg_of_Shipping Cost']

shipmode_result = pd.merge(pd.merge(pd.merge(pd.merge(pd.merge(pd.merge(pd.merge(pd.merge(pd.merge(pd.merge(pd.merge(
                            shipmode_qty,shipmode_count_order,on='Ship Mode'),shipmode_sales,on='Ship Mode'),shipmode_profit,on='Ship Mode'),shipmode_ship_cost,on='Ship Mode')
                           ,shipmode_avg_sales,on='Ship Mode'),shipmode_avg_profit,on='Ship Mode'),shipmode_avg_ship_cost,on='Ship Mode')
                           ,shipmode_avg_sales_per_qty,on='Ship Mode'),shipmode_avg_profit_per_qty,on='Ship Mode'),shipmode_avg_shipping_cost_per_qty,on='Ship Mode')
                           ,shipmode_avg_aging,on='Ship Mode')
shipmode_result
shipmode_result = shipmode_result[shipmode_result['Ship Mode'] != '45788']
shipmode_result
```
